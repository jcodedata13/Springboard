```python
df1['time_stamp'] = pd.to_datetime(df1['time_stamp'])
df1.sort_values('time_stamp',inplace=True)
df1.set_index('time_stamp',inplace=True)
```

```python
grouped1 = df1.groupby('user_id').resample('1W').sum()
```

```python
grouped1.head(30)
```

```python
grouped1['visited'].value_counts()
```

```python
grouped1 = grouped1.drop(columns='user_id')
```

```python
grouped1
```

```python
grouped1['adopted user'] = ((grouped1['visited'] >= 3) == True)
```

```python
grouped1.head(30)
```

```python
grouped2 = grouped1.groupby('user_id')['adopted user'].sum()
```

```python
len(grouped2)
```

```python
grouped2
```

```python
grouped2=grouped2.astype(bool)
```

```python
df2.info()
```

```python
df2.head(10)
```

```python
df2['object_id'].value_counts()
```

```python
df2['last_session_creation_time'] = pd.to_datetime(df2['last_session_creation_time'],unit='s')
```

```python
df2 = df2.drop(columns=['invited_by_user_id'])
```

```python
df2.head()
```

```python
df = pd.merge(grouped2,df2, left_on='user_id',right_on='object_id',how='outer')
```

```python
df.head(30)
```

```python
df.info()
```

```python
df['adopted user'].fillna(False,inplace=True)
df.drop(columns=['last_session_creation_time','name','email','creation_time'],inplace=True)
df.set_index('object_id',inplace=True)
```

```python
df.head(30)
```

```python
dummies = pd.get_dummies(df['creation_source'])
```

```python
df = df.join(dummies)
```

```python
df = df.drop(columns='creation_source')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['adopted user'],axis=1)
y = df['adopted user']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=41)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
random_grid = {'n_estimators':np.arange(1,50)}
clf = RandomForestClassifier()
clf_cv = GridSearchCV(clf,random_grid,cv=5)
clf_cv.fit(X_train,y_train)
```

```python
clf_cv.best_params_
```

```python
clf_cv.best_estimator_.feature_importances_
```

```python
X_train_columns = X_train.columns
```

```python
#Feature Importance
feature_df = pd.DataFrame({'Feature':X_train_columns,'Importance':clf_cv.best_estimator_.feature_importances_})
```

```python
feature_df.sort_values(by='Importance',ascending=False)
```
